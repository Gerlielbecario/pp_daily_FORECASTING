```python
def find_desired_hour_indices(steps, desired_hours):
    # Convertir los pasos a un objeto Series de pandas para trabajar con fechas
    steps_series = pd.Series(steps.values)

    # Obtener la hora de cada paso
    horas = steps_series.dt.components.hours

    # Encontrar los índices donde se encuentran las horas deseadas
    desired_indices = [idx for idx, hora in enumerate(horas) if hora in desired_hours]

    return desired_indices

# Horas deseadas: 00, 06, 12 y 18 UTC
desired_hours = [0, 6, 12, 18]

# Obtener los índices de las horas deseadas
desired_indices = find_desired_hour_indices(ds.step, desired_hours)



# Imprimir los índices de las horas deseadas
print(desired_indices)
```
